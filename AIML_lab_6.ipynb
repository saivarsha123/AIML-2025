```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("/content/Classified_Data.txt",
                 index_col = 0)
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
sns.pairplot(df, hue="TARGET CLASS", kind = "scatter")
```

```python
sns.pairplot(df, hue="TARGET CLASS", kind = "kde")
```

```python
l = df.columns
for i in range(len(l)-1):
  sns.boxplot(x="TARGET CLASS" ,y=l[i], data=df)
  plt.figure()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
scaled_features = sc.fit_transform(df.drop("TARGET CLASS", axis=1))
scaled_features
```

```python
df_scaled = pd.DataFrame(scaled_features, columns=df.columns[:-1])
df_scaled.head()
```

```python
sns.heatmap(df_scaled.corr(), annot=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_scaled
y = df["TARGET CLASS"]
x_train, x_test, y_train, y_test = train_test_split(scaled_features,df["TARGET CLASS"],test_size=0.20, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
```

```python
y_train_pred = knn.predict(x_train)
y_test_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print("Training Accuracy = ", accuracy_score(y_train, y_train_pred))
print("Testing Accuracy = ", accuracy_score(y_test, y_test_pred))
```

```python
cm = confusion_matrix(y_test, y_test_pred)
cm
```

```python
sns.heatmap(cm,annot = True)
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
print("Misclassification error rate:",np.mean(y_test_pred!=y_test))
```

```python
n = 60
```

```python
error_rate = []
for i in range(1,n):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  pred_i = knn.predict(x_test)
  error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,n),error_rate,color="blue",linestyle="dashed",marker="o",markerfacecolor="red",markersize=8)
plt.title("Error Rate vs. K Value", fontsize=20)
plt.xlabel("k",fontsize=15)
plt.ylabel("Error (misclassification) Rate",fontsize=15)
```

```python
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(x_train,y_train)
```

```python
print("Training Accuracy = ", accuracy_score(y_train, knn.predict(x_train)))
print("Testing Accuracy = ", accuracy_score(y_test,knn.predict(x_test)))
```
