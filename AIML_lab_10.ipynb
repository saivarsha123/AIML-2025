```python
from sklearn.metrics import pairwise_distances_argmin
```

```python
def find_clusters(X, n_clusters, rseed=2):
  rng =np.random.RandomState(rseed)
  i = rng.permutation(X.shape[0])[:n_clusters]
  centers = X[i]
  while True:
    labels = pairwise_distances_argmin(X, centers)
    new_centers = np.array([X[labels == i].mean(0)for i in range(n_clusters)])
    if np.all(centers == new_centers):
      break
    centers = new_centers
  return centers, labels
```

```python
centers, labels = find_clusters(X, 4)
plt.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis')
plt.scatter(centers[:,0], centers[:, 1], c='yellow',s=300, alpha=0.5)
```

```python
centers, labels = find_clusters(X, 8)
plt.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis')
plt.scatter(centers[:,0], centers[:, 1], c='yellow',s=300, alpha=0.5)
```

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.7, random_state=0)
X = X[:, ::-1]
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
def plot_kmeans(kmeans, X, n_clusters=4, rseed=0, ax=None):
  labels = kmeans.fit_predict(X)
  ax = ax or plt.gca() # Changed plt.gcal to plt.gca() to get the current axes
  ax.axis('equal')
  ax.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis',edgecolors='k',zorder=2)
  centers = kmeans.cluster_centers_
  radii = [cdist(X[labels == i], [center]).max() for i, center in enumerate(centers)]
  for c, r in zip(centers, radii):
    ax.add_patch(plt.Circle(c, r, fc='#CCCCCC', lw=3, alpha=0.5, zorder=1))
```

```python
kmeans = KMeans(n_clusters=4, random_state=0, n_init=10)
plot_kmeans(kmeans, X)
```
